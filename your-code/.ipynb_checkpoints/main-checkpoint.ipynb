#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

bin_sizes = [10, 30, 50]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for i, bin_size in enumerate(bin_sizes):
    ax = axes[i]
    sns.histplot(x=titanic.Age, bins=bin_size,ax=axs[i]);    
    ax.hist(titanic.Age, bins=bin_size)
    ax.set_title(f'Age (Bins: {bin_size})')
    ax.set_xlabel('Age')
    ax.set_ylabel('Freq')
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
corr = titanic.corr(method='pearson', numeric_only=True)
corr
```

#### What are the most correlated features?

```python
sns.set(style='white')     # estilo blanco

mascara = np.triu(np.ones_like(corr, dtype=bool))   # mascara

cmap = sns.diverging_palette(0, 10, as_cmap=True)   # paleta de colores


sns.heatmap(corr,          # datos
            mask=mascara,  # mascara
            cmap=cmap,     # paleta de colores
            vmax=1,           # maximo vertical
            vmin=-1,          # minimo vertical
            center=0,         # centro
            square=True,      # cuadrado el dato pintado
            linewidth=0.5,    # ancho de la linea
            
            cbar_kws={'shrink': 0.5},  # barra de color
            annot=True                 # anotacion, el numero escrito
           );
```

```python
"""
The most positive correlated variables are: Parch-SibS
"""
```
