#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,4));

x = ["Minutes Lightly Active","Minutes Fairly Active","Minutes Very Active"]

for i in range(0,3):
    sns.scatterplot(data=fitbit, x = x[i], y="Steps", ax=axs[i])
    
```

```python

fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,4));
sns.scatterplot(data=fitbit, x="Minutes Lightly Active", y="Steps", ax=axs[0]);
sns.scatterplot(data=fitbit, x="Minutes Fairly Active", y="Steps", ax=axs[1]);
sns.scatterplot(data=fitbit, x="Minutes Very Active", y="Steps", ax=axs[2]);
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
# your code here

bin_sizes = [10, 30, 50]

fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for i, bin_size in enumerate(bin_sizes):
    ax = axes[i]
    sns.histplot(x=titanic.Age, bins=bin_size,ax=axs[i]);    
    ax.hist(titanic.Age, bins=bin_size)
    ax.set_title(f'Age (Bins: {bin_size})')
    ax.set_xlabel('Age')
    ax.set_ylabel('Freq')
```

#### How does the bin size affect your plot?

```python
#By increasing the number of bins we can see more clearly how the age of the passengers is distributed.
```
