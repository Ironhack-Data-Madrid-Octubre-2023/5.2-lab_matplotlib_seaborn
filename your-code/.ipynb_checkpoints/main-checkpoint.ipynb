#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED

average_steps_by_month_encoded = fitbit.groupby('months_encoded')['steps'].mean()

average_steps_by_month_encoded
```

```python
# Utlizo un bar plot para visualizarlo:

plt.figure(figsize=(12, 4))
plt.bar(average_steps_by_month_encoded.index, average_steps_by_month_encoded.values)

plt.xticks(average_steps_by_month_encoded.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.xlabel('Month')
plt.ylabel('Average Steps')

plt.show()
```

```python
# AVERAGE STEPS BY WORK_OR_WEEKEND

average_steps_by_work_or_weekend = fitbit.groupby('work_or_weekend')['steps'].mean()
```

```python
# Utlizo un bar plot para visualizarlo:

plt.figure(figsize=(12, 4))
average_steps_by_work_or_weekend.plot(kind='bar', color=['green', 'lightblue'])

plt.xticks(rotation=0)
plt.xlabel('Day Type')
plt.ylabel('Average Steps')

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
feature_pairs = [
    ('minutes_lightly_active', 'steps'),
    ('minutes_very_active', 'steps'),
    ('minutes_sedentary', 'steps')
]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))


for i, (x_feature, y_feature) in enumerate(feature_pairs):
    
    ax = axes[i]
    ax.scatter(fitbit[x_feature], fitbit[y_feature], alpha=0.5)
    
    ax.set_xlabel(x_feature)
    ax.set_ylabel(y_feature)


plt.tight_layout()
plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
plt.style.use('classic')
plt.figure(figsize=(12, 6))

plt.hist(titanic['age'], bins=20, edgecolor='k', color='skyblue')

plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
plt.style.use('classic')

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

bins_values = [10, 20, 50]

for i, bins in enumerate(bins_values):
    ax = axes[i]
    ax.hist(titanic['age'].dropna(), bins=bins, edgecolor='k', color='skyblue')
    ax.set_xlabel("Age")
    ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

#### How does the bin size affect your plot?

```python
"""
La elección del tamaño del bin en un histograma debe alinearse con los objetivos específicos del análisis. 

Un tamaño de bin pequeño es adecuado para una exploración detallada, mientras que un tamaño de bin grande 
simplifica la visualización. 
El tamaño óptimo del bin depende del conjunto de datos y de las ideas que deseas obtener de él.
"""
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

sns.boxplot(titanic, y="age", palette="Set3")

plt.show()
```

#### What does the last plot tell you about the feature `Age`?

La línea horizontal en la caja del boxplot representa la mediana de las edades de los pasajeros. En este caso, la mediana está cerca de los 28-30 años, lo que significa que aproximadamente la mitad de los pasajeros tienen edades por encima de ese valor y la otra mitad por debajo.

La caja en el boxplot muestra el rango intercuartil, que abarca el 50% central de los datos. La parte inferior de la caja (primer cuartil, Q1) y la parte superior de la caja (tercer cuartil, Q3) indican el rango donde se encuentra la mayoría de las edades de los pasajeros.

Los puntos individuales fuera de las barras de bigotes son valores atípicos o extremos. Indican edades que son significativamente más altas o más bajas que la mayoría de los pasajeros.

Los bigotes del boxplot muestran la extensión de los datos. Se extienden 1,5 veces el IQR, lo que significa que los valores extremos se consideran más allá de ese límite.

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))

sns.violinplot(titanic, y="age", color="lightblue")

plt.ylabel("Age")
plt.xlabel("Densidad de probabilidad")

plt.show()
```

#### What additional information does the last plot provide about feature `Age`?

El violin plot muestra la distribución de la edad en forma de densidad de probabilidad. Puede observar la forma y la dispersión de la distribución de edades en el conjunto de datos del Titanic.

La línea vertical muestra la mediana de la edad en el conjunto de datos, lo cual es útil para comprender dónde se encuentra el punto medio en la distribución(y es menos sensible a valores extremos en comparación con la media).

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib

plt.figure(figsize=(12, 6))
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

```python
# Method 2 - seaborn

plt.figure(figsize=(12, 6))
sns.scatterplot(titanic, x='age', y='fare', alpha=0.5)

plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
sns.set(style="white")
plt.figure(figsize=(12, 8))

sns.heatmap(titanic.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.show()
```

#### What are the most correlated features?

Las celdas con los valores de correlación más cercanos a 1 (positivamente correlacionadas) o -1 (negativamente correlacionadas).
Es decir, los valores que están más lejanos a 0.
