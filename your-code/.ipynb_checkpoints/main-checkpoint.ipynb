#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED
ax = fitbit.groupby('Months_encoded').agg({'Steps':'median'}).plot(kind='bar',figsize=(12, 4));

months_labels = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

ax.set_xticks(range(0, 12))
ax.set_xticklabels(months_labels);
```

```python
# AVERAGE STEPS BY WORK_OR_WEEKEND
ax = fitbit.groupby('Work_or_Weekend').agg({'Steps':'median'}).plot(kind='bar',figsize=(12, 4));
ax.set_xticks([0, 1])
ax.set_xticklabels(['Weekend', 'Week']);
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

#create subplots 
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for i, feature in enumerate(features):
    ax = axes[i]
    ax.scatter(fitbit['Steps'], fitbit[feature], alpha=0.5)  
    ax.set_xlabel('Steps')
    ax.set_ylabel(feature)
    ax.set_title(f'Scatter plot: Steps vs {feature}')
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
#creamos los subplots
fig, ax = plt.subplots(1,3, figsize=(15,5))

#creamos cada hist
for i,bins in enumerate([10,20,50]): 
    ax_current = ax[i]  
    titanic['Age'].plot.hist(bins=bins, ax=ax_current)
    ax_current.set_title(f'Histogram of Age (bins={bins})')
    ax_current.set_xlabel('Age')
    ax_current.set_ylabel('Frequency');
```

#### How does the bin size affect your plot?

```python
"""
visually: the more bins, the narrower the bin
in practical terms: the bin size will change the grouping of the X-axis: the more beans, the more detailed is the distribution
"""
```

#### Plot the correlation matrix using seaborn.

```python
corr = numerical_vars.corr(method='pearson')
corr
```

```python
sns.set(style='white')     # estilo blanco

mascara = np.triu(np.ones_like(corr, dtype=bool))

cmap = sns.diverging_palette(0, 10, as_cmap=True)

sns.heatmap(corr,         
            mask=mascara,  
            cmap=cmap,     
            square=True,     
            linewidth=0.5,   
            
            cbar_kws={'shrink': 0.5},  
            annot=True                 
           );
```

#### What are the most correlated features?

```python
"""
P-Class and Fare - lo mejor la clase, lo mas caro
P-class and Age - cuando mayor, mejor la clase

"""
```
