#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4

```python
# AVERAGE STEPS BY MONTH_ENCODED

avg_steps_by_month = fitbit.groupby('Months_encoded')['Steps'].median()
avg_steps_by_month
```

```python
avg_steps_by_ww = fitbit.groupby('Work_or_Weekend')['Steps'].mean()
avg_steps_by_ww
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

#Explore plt.sca
#Explore plt.xticks

plt.sca(axes[0])
avg_steps_by_month.plot(kind='bar', color='blue')
plt.title('Median Steps by Month')
plt.xlabel('Month')
plt.ylabel('Median Steps')
plt.xticks(range(12), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])


plt.sca(axes[1])
avg_steps_by_ww.plot(kind='bar', color=['green','lightgreen'])
plt.title('Average Steps by Workday/Weekend')
plt.xlabel('Workday/Weekend')
plt.ylabel('Average Steps')
plt.xticks(range(2), ['Workday', 'Weekend'])


plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
opciones = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

plt.figure(figsize=(12, 4))

for i, opciones in enumerate(opciones, 1):
    plt.subplot(1, 3, i)
    plt.scatter(fitbit[opciones], fitbit['Steps'], alpha=0.5)
    plt.title(f'{opciones} vs Steps')
    plt.xlabel(opciones)
    plt.ylabel('Steps')


plt.tight_layout()
    
plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
plt.figure(figsize=(10, 6))
plt.hist(titanic['Age'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)

plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python

bin_sizes = [10, 20, 50]

# Create subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))


for i, bins in enumerate(bin_sizes):
    axes[i].hist(titanic['Age'].dropna(), bins=bins, color='skyblue', edgecolor='black')
    axes[i].set_title(f'Distribution of Age (Bins={bins})')
    axes[i].set_xlabel('Age')
    axes[i].set_ylabel('Frequency')
    axes[i].grid(True)

plt.tight_layout()

plt.show()

```

#### How does the bin size affect your plot?

El tamao de los intervalos (o "bins") en un histograma afecta la apariencia y la interpretacin del grfico al controlar la cantidad de barras utilizadas para representar los datos. La eleccin del tamao de los intervalos puede tener un impacto significativo en cmo se visualiza la distribucin de datos:

- Pequeos: Utilizar intervalos ms pequeos resulta en ms barras en el histograma, lo que proporciona un mayor nivel de detalle. Los intervalos ms pequeos pueden resaltar pequeas variaciones, pero pueden ocultar la tendencia general.
- Grandes: Utilizar intervalos ms grandes resulta en menos barras en el histograma, ofreciendo una vista ms general de la distribucin de los datos. Pueden identificar las caractersticas principales de la distribucin, pero podran pasar por alto variaciones a menor escala.

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# Create a box plot for the Age feature

plt.figure(figsize=(8, 6))
sns.boxplot(data=titanic, y='Age', color='skyblue')
plt.title('Summary Statistics of Age')
plt.ylabel('Age')
plt.grid(axis='y')

plt.show()
```

He elegido un 'box plot' para visualizar las estadsticas resumidas de la caracterstica "Age" (Edad) por las siguientes razones:
- estadsticos: representacin concisa de estadsticas clave, como la mediana, los cuartiles (25% y 75%), y los posibles valores atpicos (outliers). Esto permite comprender rpidamente la distribucin central de la edad y su dispersin.
- identificacin de Outliers: til para detectar observaciones inusuales en la distribucin de edades.
- interpretacin Visual Clara: El box plot es fcil de entender y proporciona una representacin visual eficaz de las estadsticas resumidas de la edad.

#### What does the last plot tell you about the feature `Age`?

    - que la mayora de las personas estn entre los 24 (aprox) y 38 aos. 

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it.

#### What additional information does the last plot provide about feature `Age`?
