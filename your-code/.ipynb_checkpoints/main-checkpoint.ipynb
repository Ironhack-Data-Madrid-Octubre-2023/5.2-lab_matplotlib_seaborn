#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

```python
# AVERAGE STEPS BY MONTH_ENCODED

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

fig = plt.figure(facecolor = 'none', figsize = (12, 6))
ax = fig.add_subplot(111, facecolor = 'none')

sns.barplot(y = fitbit['Steps'],
            x = fitbit['Months'],
            edgecolor = 'black',
            facecolor = 'none',
            order = month_order)

ax.set_xticklabels(ax.get_xticklabels(), 
                   rotation = 45, 
                   ha = 'right')

plt.show();
```

```python
# AVERAGE STEPS BY WORK_OR_WEEKEND

fig = plt.figure(facecolor = 'none', figsize = (12, 6))
ax = fig.add_subplot(111, facecolor='none')

sns.barplot(y = fitbit['Steps'],
            x = fitbit['Work_or_Weekend'],
            edgecolor = 'black',
            facecolor = 'none')

ax.set_xticklabels(ax.get_xticklabels(), 
                   rotation = 45, 
                   ha = 'right')

plt.show();
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
fig = plt.figure(facecolor = 'none', figsize = (6, 18))

x_axis = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
       
for x, i in zip(x_axis, range(1, 4)):
    name = f'ax{i}' 
    num = int(f'31{i}')
    name = fig.add_subplot(num, facecolor = 'none')
    name.scatter(fitbit[x], 
                 fitbit['Steps'],
                 color = 'grey',
                 s = 5)
    name.set_title('Frequency distribution', 
                  fontsize = 16)
    name.set_ylabel(x, 
                   fontsize = 12)
    name.set_xlabel('Steps', 
                   fontsize = 10)
    
    # Draw a lineal correlation: 
    # y = mx + b
    m, b = np.polyfit(fitbit[x], fitbit['Steps'], 1)
    name.plot(fitbit[x], m * fitbit[x] + b, color = 'orange')

plt.show();
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
fig = plt.figure(facecolor = 'none', 
                 figsize = (10, 2))

ax = fig.add_subplot(111, 
                     facecolor = 'none')

sns.boxplot(data = titanic['age'], 
            color = '#029e73',
            orient = 'h')

plt.show();
```

#### What does the last plot tell you about the feature `Age`?

```python
Q1 = titanic['age'].describe()['25%']
Q2 = titanic['age'].describe()['50%']
Q3 = titanic['age'].describe()['75%']

print(f"""
The box in the middle represent the 50% of the population,
ages between {Q1} and {Q3} years.
Q1 is the begining of the box, at left, and Q3 is the end.
The median is the value in the middle, {Q2} years.
There are also 'error bars' which represent the values for the Tukey Test:
'IQR = (Q3 - Q1)'
At the left, Q1 - 1.5*IQR and at the right, Q3 + 1.5*IQR
Finally, there are some points out of the box and the bars, those are outliers.
""")
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
df = titanic
x = 'fare'
y = 'age'
```

```python
# Method 1 - matplotlib

df = df.dropna(subset = [x, y])

# Display the correlation table between both columns
# display(df[[x, y]].corr())

m, b = np.polyfit(df[x], df[y], 1)

fig = plt.figure(facecolor='none')

ax = fig.add_subplot(111, facecolor='none')

ax.scatter(x = df[x],
           y = df[y],
           edgecolor = 'black',
           facecolor = 'none',
           s = 5)

ax.plot(df[x], 
        m * df[x] + b, 
        color = 'orange') 

plt.xlabel(x)
plt.ylabel(y)

plt.show();
```

```python
# Method 2 - seaborn

fig = plt.figure(facecolor = 'none', 
                 figsize = (5, 5))

ax = fig.add_subplot(111, 
                     facecolor = 'none')

sns.regplot(x = df[x], 
            y = df[y], 
            data = df, 
            scatter = True, 
            scatter_kws = {'s': 5},
            color = 'orange', 
            ax = ax)

plt.show();
```
