#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

```python
data = fitbit.groupby('months')['steps'].mean().reindex([calendar.month_name[i] for i in range(1, 13)])
```

```python
sns.barplot(x=data.index,
            y=data.values,
            width = 0.5)

plt.xticks(rotation=35)
plt.xlabel = "meses"

plt.savefig('months_steps.png');
```

```python
data2 = fitbit.groupby('work_or_weekend')['steps'].mean()
```

```python
plt.pie(data2, labels = ['work', 'weekend'], autopct='%1.1f%%', startangle=140)
plt.savefig('work_weekend_steps.png');
```

## Write a loop to plot 3 scatter plots of the following features:
    
- Minutes lightly active vs steps
- Minutes very active vs steps
- Minutes sedentary vs steps

```python
parejas = [('minutes_lightly_active', 'steps'),
           ('minutes_very_active', 'steps'),
           ('minutes_sedentary', 'steps')]

# Create a loop to generate scatter plots for each pair of features
for i, (x_feature, y_feature) in enumerate(parejas, start=1):
    plt.figure(figsize=(4, 4))
    
    x_data = fitbit[x_feature]
    y_data = fitbit[y_feature]
    
    # Create the scatter plot
    plt.scatter(x_data, y_data, alpha=0.5)
    
    plt.savefig(f'scatter{i}.png');
```

#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
titanic.age.describe()
```

```python
mean_age = titanic['age'].mean()
mode_age = titanic['age'].mode().values[0]
median_age = titanic['age'].median()

sns.histplot(data=titanic, x='age', color='w', bins=40)

plt.ylabel('recuento personas')

plt.axvline(mean_age, color='r', linestyle='dashed', linewidth=4)
plt.axvline(mode_age, color='g', linestyle='dashed', linewidth=4)
plt.axvline(median_age, color='y', linestyle='dashed', linewidth=4)

plt.savefig('titanic_age_desc.png');
```

TambiÃ©n vemos que la moda se encuentra alrededor de los 24-25 aÃ±os y que la mediana esta sobre los 29, lo que nos indica que hay el mismo nÃºmero de pasajeros menores de 29 que mayores.

#### Use seaborn to plot the distribution of `Age` based on the `Gender`.
**Hint**: Use Facetgrid.

```python
g = sns.FacetGrid(titanic, col="gender", margin_titles=True, height=4)

g.map(plt.hist, "age", bins=20, color="skyblue")

# Add labels and titles
g.set_axis_labels("age", "count")
g.set_titles(col_template="gender = {col_name}")
plt.savefig('titanic_age_gen.png');
```
