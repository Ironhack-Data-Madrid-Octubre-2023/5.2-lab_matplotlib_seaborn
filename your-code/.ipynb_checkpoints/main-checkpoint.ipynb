#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED
fig, ax1 = plt.subplots(figsize=(12, 4))
plt.sca(ax1)
average_steps_by_month = fitbit.groupby('months_encoded')['steps'].median()
average_steps_by_month.plot(kind='bar', color='skyblue')
plt.savefig('./figures/average_steps_by_month.png')
plt.show();
```

```python
# AVERAGE STEPS BY WORK_OR_WEEKEND
fig, ax2 = plt.subplots(figsize=(12, 4))
plt.sca(ax2)
average_steps_by_day = fitbit.groupby('work_or_weekend')['steps'].mean()
average_steps_by_day.plot(kind='bar', color='lightcoral')
plt.xticks([0, 1], ['Workday', 'Weekend'], rotation=0)
plt.ylabel('Average Steps')
plt.savefig('./figures/average_steps_by_day.png')
plt.show();
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ["minutes lightly active", "minutes very active", "minutes sedentary"]
colors = ['blue', 'green', 'red']
for i, feature in enumerate(features):
    plt.figure(figsize=(8, 6))  # Tamaño de la figura
    plt.scatter(fitbit[feature], fitbit['steps'], color=colors[i], alpha=0.5)
    plt.title(f'Scatter Plot: {feature} vs Steps')
    plt.xlabel(feature)
    plt.ylabel('Steps')
    plt.grid(True)
    plt.savefig(f'./figures/{feature}.png')
    plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
plt.figure(figsize=(12, 6))  # Set the figure size
plt.hist(titanic['age'], bins=20, color='skyblue')  # Adjust the number of bins as needed
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show();
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

bin_values = [10, 20, 50]
for i, bins in enumerate(bin_values):
    ax = axes[i]
    ax.hist(titanic['age'], bins=bins, color='skyblue')
    ax.set_title(f'Distribution of Age (Bins = {bins})')
    ax.set_xlabel('Age')
    ax.set_ylabel('Frequency')
    ax.grid(True)

plt.tight_layout()
plt.show();
```

#### How does the bin size affect your plot?

```python
"""
Divide el gráfico en más barras, o sea, se ve la distribución de la edad agrupada en rangos más pequeños cuantos más números de bins haya.
"""
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
n_male = len(titanic[titanic['gender'] == 'male'])
n_female = len(titanic[titanic['gender'] == 'female'])

plt.figure(figsize=(8, 6))
plt.pie([n_male, n_female], labels=['Male', 'Female'], autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()
```

```python
# Method 2 - seaborn
sns.barplot(x='gender', y='count', data=titanic.groupby('gender').size().reset_index(name='count'))
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
fare = titanic['fare']
age = titanic['age']

plt.scatter(age, fare)

plt.title('Relación entre Fare y Age')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

```python
# Method 2 - seaborn
sns.scatterplot(x='age', y='fare', data=titanic)

plt.title('Relación entre Fare y Age')
plt.xlabel('Age')
plt.ylabel('Fare')

plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
correlations = titanic.corr()

sns.heatmap(correlations, 
        xticklabels=correlations.columns,
        yticklabels=correlations.columns,
        cmap="coolwarm")
```
