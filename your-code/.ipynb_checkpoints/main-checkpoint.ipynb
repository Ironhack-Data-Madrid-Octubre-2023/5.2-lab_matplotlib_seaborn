#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays?
**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# AVERAGE STEPS BY MONTH_ENCODED

average_steps_by_month = fitbit.groupby('Months_encoded')['Steps'].mean()

fig, ax = plt.subplots(figsize=(12, 4))
average_steps_by_month.plot(kind='bar', ax=ax, color='skyblue')

ax.set_xlabel('Month')
ax.set_ylabel('Average Steps')
ax.set_title('Average Steps by Month')

plt.show()
```

```python
# AVERAGE STEPS BY WORK_OR_WEEKEND

average_steps_by_work_or_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].mean()


fig, ax = plt.subplots(figsize=(12, 4))
average_steps_by_work_or_weekend.plot(kind='bar', ax=ax, color=['lightcoral', 'lightgreen'], alpha=0.7)

ax.set_xlabel('Workday or Weekend')
ax.set_ylabel('Average Steps')
ax.set_title('Average Steps by Workday or Weekend')

ax.set_xticklabels(['Workday', 'Weekend'], rotation=0)

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for feature in features:
    plt.figure(figsize=(8, 6))
    plt.scatter(fitbit[feature], fitbit['Steps'], alpha=0.5)
    

    plt.xlabel(feature)
    plt.ylabel('Steps')
    plt.title(f'Scatter Plot of {feature} vs Steps')
    
    plt.show()
```

#### Use the right visulalization to show the distribution of column `Age`.

```python
plt.hist(titanic['Age'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

#### Use subplots and plot the distribution of the `Age`  with bins equal to 10, 20 and 50.

```python
bins = [10, 20, 50]

for i, num_bins in enumerate(bins):
    plt.subplot(1, len(bins), i + 1)
    plt.hist(titanic['Age'], bins=num_bins, color='skyblue', edgecolor='black')
    plt.title(f'Bins = {num_bins}')
    plt.xlabel('Age')
    plt.ylabel('Frequency')

plt.suptitle('Distribution of Age with Different Bins')
plt.tight_layout()
plt.show()
```

#### Use seaborn to show the distribution of column `Age`.

```python
sns.set_style("whitegrid")

sns.histplot(titanic['Age'].dropna(), bins=20, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')

# Display the plot
plt.show()
```

#### Use the right plot to visualize column `Gender`. There are 2 ways of doing it. Do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
gender_counts = titanic['Gender'].value_counts()

plt.figure(figsize=(6, 4))
plt.bar(gender_counts.index, gender_counts.values, color=['blue', 'pink'])
plt.title('Gender Distribution (Matplotlib)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Method 2 - seaborn
sns.set_style("whitegrid")
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='Gender', palette=['blue', 'pink'])
plt.title('Gender Distribution (Seaborn)')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

#### Use the right plot to visualize the column `Pclass`.

```python
pclass_counts = titanic['Pclass'].value_counts().sort_index()

plt.figure(figsize=(6, 4))
plt.bar(pclass_counts.index, pclass_counts.values, color='skyblue')
plt.title('Passenger Class Distribution (Matplotlib)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(pclass_counts.index, labels=['1st Class', '2nd Class', '3rd Class'])
plt.show()
```

#### We suspect that there is a linear relationship between `Fare` and `Age`. Use the right plot to show the relationship between these 2 features. There are 2 ways, please do it both ways.
**Hint**: Use matplotlib and seaborn.

```python
# Method 1 - matplotlib
plt.figure(figsize=(8, 6))
plt.scatter(titanic['Age'], titanic['Fare'], color='blue', alpha=0.5)
plt.title('Scatter Plot of Fare vs. Age (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
# Method 2 - seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=titanic, x='Age', y='Fare', color='blue', alpha=0.5)
plt.title('Scatter Plot of Fare vs. Age (Seaborn)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

#### Plot the correlation matrix using seaborn.

```python
numerical_columns = titanic.select_dtypes(include='number')
correlation_matrix = numerical_columns.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

#### Use seaborn to plot the distribution of `Age` based on the `Gender`.
**Hint**: Use Facetgrid.

```python
g = sns.FacetGrid(titanic, col="Gender", height=4, aspect=1)
g.map(sns.histplot, "Age", bins=20, kde=True, color='blue')
g.set_axis_labels("Age", "Count")
g.set_titles(col_template="Gender = {col_name}")
g.fig.suptitle("Age Distribution by Gender", y=1.05)

plt.tight_layout()
plt.show()
```
